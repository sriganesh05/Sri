# **Handle Outliers**


```python
df_sales_monthly
```

```python
df_sales_monthly.head()
```

```python
df_sales_monthly.describe()
```

```python
df_sales_monthly[df_sales_monthly.Qty<0]
```

```python
df_sales_monthly.drop(df_sales_monthly[df_sales_monthly["Qty"] < 0].index, inplace=True)

df_sales_monthly
```

```python

def remove_outliers(df_sales_monthly, net_sales_amount):
    Q1 = df_sales_monthly[net_sales_amount].quantile(0.25)
    Q3 = df_sales_monthly[net_sales_amount].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_filtered = df_sales_monthly[(df_sales_monthly[net_sales_amount] >= lower_bound) & (df_sales_monthly[net_sales_amount] <= upper_bound)]

    return df_filtered

df_sales_monthly_cleaned = remove_outliers(df_sales_monthly, 'net_sales_amount')

df_sales_monthly_cleaned.to_csv("fact_sales_monthly_cleaned.csv", index=False)

print("Outliers removed. Cleaned dataset saved as 'fact_sales_monthly_cleaned.csv'.")
```

## **Visualization**

First we need to convert 'date' column to date time format

```python
df_sales_monthly_cleaned["date"] = pd.to_datetime(df_sales_monthly_cleaned["date"])

```

```python
df_sales_monthly_cleaned['year'] = df_sales_monthly_cleaned['date'].dt.year
yearly_sales = df_sales_monthly_cleaned.groupby('year')['net_sales_amount'].sum()
```

```python
yearly_sales = df_sales_monthly_cleaned.groupby(["year"])["net_sales_amount"].sum().reset_index()
yearly_sales
```

```python
plt.figure(figsize=(10, 5))
plt.bar(yearly_sales['year'], yearly_sales['net_sales_amount'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Total Net Sales Amount')
plt.title('Yearly Total Net Sales')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
df_sales_monthly_cleaned['date'] = pd.to_datetime(df_sales_monthly_cleaned['date'], dayfirst=True, errors='coerce')
```

```python
product_sales = df_sales_monthly_cleaned.groupby('product_code')['net_sales_amount'].sum().sort_values(ascending=False)
product_sales
```

```python
plt.figure(figsize=(12, 6))
product_sales.head(10).plot(kind='line', color='skyblue')
plt.xlabel('Product Code')
plt.ylabel('Total Net Sales Amount')
plt.title('Top 10 Products by Total Net Sales')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
correlation_matrix = df_sales_monthly_cleaned[['customer_code', 'Qty', 'net_sales_amount']].corr()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', linewidths=0.5, fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
